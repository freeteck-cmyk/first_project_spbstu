```python
def check_input(array_element):
    if isinstance(array_element, str):# Проверяем, является ли элемент массива строкой
        return True
    else:
        return False

def get_substring(s: str) -> str: # на вход принимаем строку и возвращаем строку
    n = len(s)
    if n == 0:
        return "" # Guard Clause - если пустой или некорректный ввод. Быстрый выход

    current_start = 0
    used_chars = {}
    longest = ""

    for i in range(n): # Используем метод скользящего окна со словарем
        char = s[i]
        if char in used_chars and used_chars[char] >= current_start:
            current_start = used_chars[char] + 1
        used_chars[char] = i
        current_sub = s[current_start:i+1]

        if len(current_sub) > len(longest):
            longest = current_sub
    return longest

def main(data_list):
    max_str = ""
    target_index = -1

    for i, item in enumerate(data_list):
        if check_input(item):
            if len(item) > len(max_str):
                max_str = item
                target_index = i

    if target_index == -1:
        result_index = -1
        result_sub = ""
    else:
        result_index = target_index
        result_sub = get_substring(max_str)
    print(result_index)
    print(result_sub)

data = [[0, 4, 8], 'abca', 7, 'aaaaa']
main(data)
```
